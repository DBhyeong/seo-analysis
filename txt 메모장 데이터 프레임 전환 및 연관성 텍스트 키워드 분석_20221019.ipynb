```python
import pandas as pd

f = open(r"C:\Users\user\Desktop\test.txt", 'r',encoding='UTF8')

#data = f.readlines() #read로 말고 마침표로 한다.

data = f.read()

data = data.split('.')
column_name = ['text'] #컬럼이름

data_split = [x.strip() for x in data] #readline에 따른 리스트화

df = pd.DataFrame(data_split, columns = column_name)  #메모장을 데이터 프레임 전환

#df

```

```python
import re

def text_cleaning(text) :
    hangul = re.compile('[^ ㄱ-ㅣ가-힣+]') #한글만 가져오기
    result = hangul.sub('', text)
    return result
```

```python
df['ko_text'] = df['text'].apply(lambda x :text_cleaning(x))

df.head()
```

```python
from konlpy.tag import Okt
from collections import Counter

korean_stopwords_path = r"C:\Users\user\Desktop\stopwords.txt"
with open(korean_stopwords_path, encoding='utf-8') as f :
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]


def get_nouns(x) :
    nouns_tagger = Okt()
    nouns = nouns_tagger.nouns(x)
        
    nouns = [noun for noun in nouns if len(noun) > 1]     #한글자 키워드를 제거
    
    nouns = [noun for noun in nouns if noun not in stopwords] #불용어를 제거

    return nouns
    
```

```python
df['nouns'] = df['ko_text'].apply(lambda x : get_nouns(x))

print(df.shape)
df.head()
```

```python
from apyori import apriori

#연관분석 연습

#장바구니 형태의 데이터 (트랜잭선 데이터) 를 생성한다.

transactions = [
    
['손흥민', '시소코'],
['손흥민', '케인'],
['손흥민', '케인', '포체티노']
                
]


#연관분석 실행


results = list(apriori(transactions))
#for result in results :
#    print(result)
```

```python
#연관 키워드 분석

transactions = df['nouns'].tolist()
transactions = [transaction for transaction in transactions if transaction] #공백 문자열 방지

#print(transactions)
```

```python
#연관 분석 수행

results = list(apriori(transactions, 
                       min_support = 0.02,
                       #min_confidence = 0.2,
                       #min_lift = 5,
                       max_length = 2
                      ))
#print(results)
```

```python
#데이터 프레임 형태로 정리

columns = ['source', 'target', 'support']

network_df = pd.DataFrame(columns=columns)

#규칙의 조건절을 source, 결과절은 target, 지지도를 support 라는 데이터 프레임의 피처로 반환

for result in results :
    if len(result.items) == 2 :
        items = [x for x in result.items]
        row = [items[0], items[1], result.support]
        series = pd.Series(row, index= network_df.columns)
        network_df = network_df.append(series, ignore_index=True)
network_df
```

```python
#말뭉치 추출

txt_corpus = "".join(df['ko_text'].tolist())
#print(txt_corpus)
```

```python
#명사 키워드 추출

from konlpy.tag import Okt
from collections import  Counter


nouns_tagger = Okt() #명사 키워드 추출
nouns = nouns_tagger.nouns(txt_corpus)
count = Counter(nouns)

#한글자 키워드 제거


remove_char_counter = Counter({x : count[x] for x in count if len(x) >1 })
#print(remove_char_counter)
```

```python
#단어 빈도 점수 추가

#키워드와 키워드 빈도 점수를 'node', 'nodesize' 라는 데이터 프레임의 피처로 생성
node_df = pd.DataFrame(remove_char_counter.items(), columns=['node', 'nodesize'])
#node_df = node_df[node_df['nodesize']>=50] #시각화 편의를 위해 nodesize 50 이하는 제거

node_df.head()
```

```python
import networkx as nx
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm #폰트설정
import warnings

from matplotlib import font_manager, rc

warnings.filterwarnings('ignore')

plt.figure(figsize=(25,25))

#networkx 그래프 객체를 생성

G = nx.Graph()

# node_df 의 키워드 빈도수를 데이터로 하여, 네트워크 그래프의 '노드' 역할을 하는 원을 생성

for index, row in node_df.iterrows() :
    G.add_node(row['node'], nodesize= row['nodesize'])
    
# noetwork_df 의 연관 분석 데이터를 기반으로 네트워크 그래프의 '관계' 역할을 하는 선을 생성

for index, row in network_df.iterrows () :
    G.add_weighted_edges_from([(row['source'], row['target'], row['support'])])

#그래프 디자인과 관련된 파라미터를 설정

pos = nx.spring_layout(G, k=0.1, iterations = 25)
#sizes = [G.node[node]['nodesize']*25 for node in G]

sizes = [G.nodes[node]['nodesize'] * 25 for node in G]

nx.draw(G, pos=pos, node_size=sizes)


# 폰트 설정을 위한 font_manager import

font_path = r'C:\Users\user\AppData\Local\Microsoft\Windows\Fonts\D2Coding-Ver1.3.2-20180524-all.ttc'

# 폰트 설정
#fm._rebuild()		# 1회에 한해 실행해준다. (폰트 새로고침, 여러번 해줘도 관계는 없다.)
#font_fname = 'C:/Windows/Fonts/맑은 고딕/malgun.ttf'      # 여기서 폰트는 C:/Windows/Fonts를 참고해도 좋다.
#font_fname = font_fname.replace("\\", '/')
#fontprop = fm.FontProperties(fname=font_fname, size=18).get_name()
font = font_manager.FontProperties(fname=font_path, size=15).get_name() #넣기


nx.draw_networkx_labels(G, pos=pos, font_family=font, font_size=25)


#그래프를 출력

ax =plt.gca()

plt.show()
```
