```python
import urllib.request
import json
from pandas.io.json import json_normalize
import hashlib
import hmac
import base64
import pandas as pd
import os
import requests
from    bs4      import BeautifulSoup
import time
import warnings

warnings.filterwarnings('ignore')

f = open('C:/Users/user/raw/kws/kw_serp_list.txt', 'r', encoding='utf-8-sig')
keywords = f.readlines() 


naver_kw_serp_section_data_view_pc = []   #담을 내용
df = pd.DataFrame(naver_kw_serp_section_data_view_pc, columns=['keyword','blog', 'visibility'])  #컬럼지정

for keyword in keywords :
    keyword = keyword.replace("\n", "")
    url ='https://search.naver.com/search.naver?where=view&sm=tab_jum&query=' + str(keyword)
    raw = requests.get(url)  #협찬
    html = BeautifulSoup(raw.text, 'html.parser') #긁기


    headers = { 'Accept-Language' : 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7,zh-TW;q=0.6,zh;q=0.5',
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36',
                'Accept-Encoding': 'gzip'
    }

    raw = requests.get(url=url, headers=headers, verify =False)
    html = BeautifulSoup(raw.text, 'html.parser')
    
    results = html.select('#main_pack > section')  #블로그 검색 섹션 보기
    
    try : #예외처리 추가
        results_2 = results[0].find_all('li')  #li 리스트 가져오기

        #print(results_2)
        text = str(results_2[:5]) #상위 5위까지 텍스트화

        ableclass = 'ablewithus'
        viewracle = 'viewracle'
        viewracle21 = 'viewracle21'

        #print(html.prettify())

        if text.find(ableclass) >= 0 :  #해당 키워드가 노출되는지 여부 (ableclass)

            #print(str(keyword) + ',' + str(ableclass) + ',' + 'O')
            new_data = {
                'keyword' : keyword,
                'blog' :  ableclass,
                'visibility' :  'O'
            }
            df = df.append(new_data, ignore_index=True)
            
        else :

            #print(str(keyword) + ',' + str(ableclass) + ',' + 'X')
            new_data = {
                'keyword' : keyword,
                'blog' :  ableclass,
                'visibility' :  'X'
            }
            df = df.append(new_data, ignore_index=True)


        if text.find(viewracle) >= 0 :  #해당 키워드가 노출되는지 여부 (viewracle)

            #print(str(keyword) + ',' + str(viewracle) + ',' + 'O')

            new_data = {
                'keyword' : keyword,
                'blog' :  viewracle,
                'visibility' :  'O'
            }
            df = df.append(new_data, ignore_index=True)


        elif text.find(viewracle21) >= 0 :  #해당 키워드가 노출되는지 여부 (viewracle)
            #print(str(keyword) + ',' + str(viewracle) + ',' + 'O')

            new_data = {
                'keyword' : keyword,
                'blog' :  viewracle,
                'visibility' :  'O'
            }
            df = df.append(new_data, ignore_index=True)

        else :                  #해당 키워드가 노출되는지 여부 (viewracle)
            #print(str(keyword) + ',' + str(viewracle) + ',' + 'X')
            new_data = {
                'keyword' : keyword,
                'blog' :  viewracle,
                'visibility' :  'X'
            }

            df = df.append(new_data, ignore_index=True)
    except :
        new_data = {
                'keyword' : keyword,
                'blog' :  ableclass,
                'visibility' :  'X'
            }
        df = df.append(new_data, ignore_index=True)
    
        new_data = {
                'keyword' : keyword,
                'blog' :  viewracle,
                'visibility' :  'X'
            }
        df = df.append(new_data, ignore_index=True)
    
    
    print(str(keyword) + "의 검색결과를 완료했습니다.")

    time.sleep(3) #1초
    
data = df.pivot_table(index='keyword', columns='blog', values='visibility' , aggfunc='max')
data.to_csv('C:/Users/user/raw/kws/naver_kw_serp_section_data_view_pc.csv', encoding ='utf-8-sig')  #최종 엑셀 파일로 보내기

print("최종완료")
```
